## Flipping over Y axis

Reflecting the video over the middle so that the data is being shown in both directions

```python
import pandas as pd
import os
import glob
import sys

# ==========================================
# CONFIGURATION 
# ==========================================
# Input folders
train_folder_path = 'Not_Seperated/Clips_Split_80_20/Train_clips' 
test_folder_path = 'Not_Seperated/Clips_Split_80_20/Test_clips'

# Output directories
train_out_dir = 'Not_Seperated/Clips_Split_80_20/Train_clips_augmented'
test_out_dir = 'Not_Seperated/Clips_Split_80_20/Test_clips_augmented'

# ==========================================
# DIRECTORY VALIDATION
# ==========================================
# Don't create folders; check if they already exist
directories_to_check = [train_folder_path, test_folder_path, train_out_dir, test_out_dir]
folders_missing = False

for directory in directories_to_check:
    if not os.path.exists(directory):
        print(f"❌ ERROR: Cannot find folder -> '{directory}'")
        folders_missing = True

# Stop the script if any folders are missing
if folders_missing:
    print("\nPlease create the missing folders above or fix the paths before running again.")
    # Exits the cell early
    sys.exit()

print("✅ All folders found. Starting augmentation...\n")

# The Left/Right pairs based on your 21-point mapping
left_right_pairs = [
    (1, 2),   # Eyes
    (3, 4),   # Ears
    (5, 6),   # Mouth
    (7, 8),   # Shoulders
    (9, 10),  # Elbows
    (11, 12), # Wrists
    (13, 14), # Hips
    (15, 16), # Knees
    (17, 18), # Ankles
    (19, 20)  # Heels
]

# ==========================================
# 1. PROCESS TRAINING DATA
# ==========================================
print(f"Processing Training Data from: {train_folder_path}")
# Find all CSV files in the training folder
train_files = glob.glob(os.path.join(train_folder_path, '*.csv'))

if not train_files:
    print(f"  ⚠️ No CSV files found in {train_folder_path}")
else:
    for file_path in train_files:
        df = pd.read_csv(file_path)

        # Identify x columns
        x_columns = [col for col in df.columns if col.startswith('x') and col[1:].isdigit()]

        # Create flipped dataframe and apply 1.0 - x
        df_flipped = df.copy()
        df_flipped[x_columns] = 1.0 - df_flipped[x_columns]

        # Swap left and right points to fix anatomy
        for left_idx, right_idx in left_right_pairs:
            for axis in ['x', 'y', 'z', 'v']:
                left_col = f'{axis}{left_idx}'
                right_col = f'{axis}{right_idx}'
                
                if left_col in df_flipped.columns and right_col in df_flipped.columns:
                    df_flipped[[left_col, right_col]] = df_flipped[[right_col, left_col]]

        # Save both files separately
        filename = os.path.basename(file_path)
        name, ext = os.path.splitext(filename) # Splits 'video1.csv' into 'video1' and '.csv'
        
        orig_out_path = os.path.join(train_out_dir, filename)
        flipped_out_path = os.path.join(train_out_dir, f"{name}_flipped{ext}")
        
        df.to_csv(orig_out_path, index=False)
        df_flipped.to_csv(flipped_out_path, index=False)
        
        print(f"  -> Saved '{filename}' and '{name}_flipped{ext}'")


# ==========================================
# 2. PROCESS TESTING DATA
# ==========================================
print(f"\nProcessing Testing Data from: {test_folder_path}")
# Find all CSV files in the testing folder
test_files = glob.glob(os.path.join(test_folder_path, '*.csv'))

if not test_files:
    print(f"  ⚠️ No CSV files found in {test_folder_path}")
else:
    for file_path in test_files:
        df = pd.read_csv(file_path)

        # Identify x columns
        x_columns = [col for col in df.columns if col.startswith('x') and col[1:].isdigit()]

        # Create flipped dataframe and apply 1.0 - x
        df_flipped = df.copy()
        df_flipped[x_columns] = 1.0 - df_flipped[x_columns]

        # Swap left and right points to fix anatomy
        for left_idx, right_idx in left_right_pairs:
            for axis in ['x', 'y', 'z', 'v']:
                left_col = f'{axis}{left_idx}'
                right_col = f'{axis}{right_idx}'
                
                if left_col in df_flipped.columns and right_col in df_flipped.columns:
                    df_flipped[[left_col, right_col]] = df_flipped[[right_col, left_col]]

        # Save both files separately
        filename = os.path.basename(file_path)
        name, ext = os.path.splitext(filename) # Splits 'video1.csv' into 'video1' and '.csv'
        
        orig_out_path = os.path.join(test_out_dir, filename)
        flipped_out_path = os.path.join(test_out_dir, f"{name}_flipped{ext}")
        
        df.to_csv(orig_out_path, index=False)
        df_flipped.to_csv(flipped_out_path, index=False)
        
        print(f"  -> Saved '{filename}' and '{name}_flipped{ext}'")

print("\n🎉 Data augmentation complete!")
```

## Shifting all points

```python
import pandas as pd
import os
import glob
import sys
import random

# ==========================================
# CONFIGURATION
# ==========================================
# Both input and output are the same augmented directories
train_dir = 'Not_Seperated/Clips_Split_80_20/Train_clips_augmented'
test_dir = 'Not_Seperated/Clips_Split_80_20/Test_clips_augmented'

# ==========================================
# DIRECTORY VALIDATION
# ==========================================
directories_to_check = [train_dir, test_dir]
folders_missing = False

for directory in directories_to_check:
    if not os.path.exists(directory):
        print(f"❌ ERROR: Cannot find folder -> '{directory}'")
        folders_missing = True

if folders_missing:
    print("\nPlease fix the paths before running again.")
    sys.exit()

print("✅ Folders found. Starting translation augmentation...\n")

def apply_translations(folder_path, folder_label):
    print(f"Processing {folder_label} Data in: {folder_path}")
    files = glob.glob(os.path.join(folder_path, '*.csv'))

    if not files:
        print(f"  ⚠️ No CSV files found in {folder_path}")
        return

    for file_path in files:
        filename = os.path.basename(file_path)
        
        # SAFETY CHECK: Skip files that have already been translated
        # This prevents the script from infinitely translating already-translated files
        # if you ever run this cell more than once!
        if '_translated_' in filename:
            continue

        df = pd.read_csv(file_path)
        name, ext = os.path.splitext(filename)

        # Identify x and y columns
        x_columns = [col for col in df.columns if col.startswith('x') and col[1:].isdigit()]
        y_columns = [col for col in df.columns if col.startswith('y') and col[1:].isdigit()]

        # --------------------------------------------------
        # TRANSLATE VIDEO (Shift X and Y)
        # --------------------------------------------------
        # Find global min and max for all x and y points across the entire video
        min_x = df[x_columns].min().min()
        max_x = df[x_columns].max().max()
        min_y = df[y_columns].min().min()
        max_y = df[y_columns].max().max()

        # Calculate maximum allowed shifts so points stay strictly within [0.0, 1.0]
        max_shift_x_pos = 1.0 - max_x
        max_shift_x_neg = -min_x
        max_shift_y_pos = 1.0 - max_y
        max_shift_y_neg = -min_y

        # Generate 10 random translation combinations (x_shift, y_shift)
        all_translations = []
        for _ in range(10):
            # random.uniform picks a random float between the negative and positive limits
            shift_x = random.uniform(max_shift_x_neg, max_shift_x_pos)
            shift_y = random.uniform(max_shift_y_neg, max_shift_y_pos)
            all_translations.append((shift_x, shift_y))

        # Sample 7 random translations from the 10 generated
        sampled_translations = random.sample(all_translations, 7)

        # Apply and save the 7 translated videos
        for i, (sx, sy) in enumerate(sampled_translations, start=1):
            df_translated = df.copy()
            
            # Add the shifts to all X and Y columns
            df_translated[x_columns] = df_translated[x_columns] + sx
            df_translated[y_columns] = df_translated[y_columns] + sy
            
            # Save back into the same folder
            trans_out_path = os.path.join(folder_path, f"{name}_translated_{i}{ext}")
            df_translated.to_csv(trans_out_path, index=False)

        print(f"  -> Processed '{filename}': Saved 7 translated copies.")

# ==========================================
# EXECUTE
# ==========================================
apply_translations(train_dir, "Training")
print("-" * 50)
apply_translations(test_dir, "Testing")

print("\n🎉 Translation augmentation complete!")
```
